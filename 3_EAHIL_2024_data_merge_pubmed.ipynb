```python
import pandas as pd

# sources
myfolder_temp = 'data/temp/2024/'
myfolder_results = 'results/2024/'
aou_data = myfolder_results + 'aou_biomed_2015_2022.tsv'
aou_pmids = myfolder_temp + 'aou_pmids.tsv'
aou_dois = myfolder_temp + 'aou_dois.tsv'
export_pubmed1 = 'data/sources/pubmed_export_unige_hug_2015_2017_20240529.csv'
export_pubmed2 = 'data/sources/pubmed_export_unige_hug_2018_2020_20240529.csv'
export_pubmed3 = 'data/sources/pubmed_export_unige_hug_2021_2022_20240529.csv'
export_pubmed_data = 'data/sources/pubmed_export_unige_hug_with_associated_data_2015_2022_20240529.csv'

# display all columns
pd.set_option('display.max_columns', None)
```

## Import AoU data

```python
aou = pd.read_csv(aou_data, dtype={'date': 'int'}, encoding='utf-8', header=0, sep='\t')
aou
```

```python
aou_dois = pd.read_csv(aou_dois, dtype={'date': 'int'}, encoding='utf-8', header=0, sep='\t')
aou_dois
```

```python
aou_pmids = pd.read_csv(aou_pmids, dtype={'date': 'int'}, encoding='utf-8', header=0, sep='\t')
aou_pmids
```

```python
# DOIs - normalize before merge
aou_dois = aou_dois.rename(columns={'doi' : 'DOI_AOU'})
aou_dois['DOI'] = aou_dois['DOI_AOU'].str.upper()
aou_dois
```

```python
# check duplicates (AoU refs with several DOIs)
aou_dois.loc[aou_dois.duplicated(subset='id', keep=False)].sort_values(by='id')
```

```python
# check duplicates (AoU refs with the same DOI)
aou_dois.loc[aou_dois.duplicated(subset='DOI', keep=False)].sort_values(by='DOI')
```

```python
# PMIDs
aou_pmids = aou_pmids.rename(columns={'pmid' : 'PMID'})
aou_pmids
```

```python
# check duplicates (AoU refs with several PMIDs)
aou_pmids.loc[aou_pmids.duplicated(subset='id', keep=False)].sort_values(by='id')
```

```python
# check duplicates (AoU refs with the same PMID)
aou_pmids.loc[aou_pmids.duplicated(subset='PMID', keep=False)].sort_values(by='PMID')
```

```python
aou_pmids.dtypes
```

```python
# export duplicates
aou_dois.loc[aou_dois.duplicated(subset='id', keep=False)].sort_values(by='id').to_csv(myfolder_temp + 'aou_dois_duplicated_ids.tsv', sep='\t', index=False)
aou_dois.loc[aou_dois.duplicated(subset='DOI', keep=False)].sort_values(by='DOI').to_csv(myfolder_temp + 'aou_dois_duplicated_dois.tsv', sep='\t', index=False)
aou_pmids.loc[aou_pmids.duplicated(subset='id', keep=False)].sort_values(by='id').to_csv(myfolder_temp + 'aou_pmids_duplicated_ids.tsv', sep='\t', index=False)
aou_pmids.loc[aou_pmids.duplicated(subset='PMID', keep=False)].sort_values(by='PMID').to_csv(myfolder_temp + 'aou_pmids_duplicated_pmids.tsv', sep='\t', index=False)
```

```python
# depdup
aou_dois = aou_dois.drop_duplicates(subset='id', keep='first')
aou_dois = aou_dois.drop_duplicates(subset='DOI', keep='first')
aou_pmids = aou_pmids.drop_duplicates(subset='id', keep='first')
aou_pmids = aou_pmids.drop_duplicates(subset='PMID', keep='first')
```

```python
# complete AoU data
aou = aou.merge(aou_dois, on='id', how='left')
aou = aou.merge(aou_pmids, on='id', how='left')
aou
```

```python
# convert nans
aou['PMID'] = aou['PMID'].fillna(0)
aou['PMID'] = aou['PMID'].astype(int)
aou['DOI'] = aou['DOI'].fillna('')
aou
```

```python
# exports
aou.to_csv(myfolder_results + 'aou_biomed_2015_2022_dois_pmids.tsv', sep='\t', index=False)
aou.to_excel(myfolder_results + 'aou_biomed_2015_2022_dois_pmids.xlsx', index=False)
```

```python
pubmed_merged_by_pmid = df.loc[df['MERGED_BY_PMID'] == 1]
pubmed_merged_by_pmid['PMID_AOU'] = pubmed_merged_by_pmid['PMID']
pubmed_merged_by_pmid = pubmed_merged_by_pmid.rename(columns={'PMID' : 'PMID_PUBMED'})
pubmed_merged_by_pmid
```

```python
pubmed_not_merged_by_pmid = df.loc[(df['MERGED_BY_PMID'] == 0) & (df['PUBMED'] == 1)]
pubmed_not_merged_by_pmid = pubmed_not_merged_by_pmid.rename(columns={'PMID' : 'PMID_PUBMED'})
pubmed_not_merged_by_pmid
```
